### Cleaning Data

```python
lx_houses = pd.read_csv("lisbon.csv")
lis_list = list(lx_houses['neighbourhood'].str.strip().values + ', Lisboa')
```

```python
bcn_houses = pd.read_csv("barcelona.csv")
bcn_list = list(bcn_houses['neighbourhood'].str.strip().values + ', Barcelona')
```

### LISBON - add data to table

```python
final = []
for loc in lis_list:
    if loc == 'Caparica, Lisboa':
        loc = 'Caparica, Almada'
        
    address2 = "Heden Santa Apolonia Doca Jardim do Tabaco Terminal de Cruzeiros de Lisboa Edifício NE, 1º Piso, 1100-651 Lisboa"

    now = datetime.now()
    try:
        directions_result = gmaps.directions(loc,
                                            address2,
                                            mode="transit",
                                            departure_time=now)
        
        # Get distance
        result_distance = directions_result[0]['legs'][0]['distance']['text']

        # Get duration of travelling with public transportation
        result_time = directions_result[0]['legs'][0]['duration']['text']
        final.append({'neighbourhood' : loc, 'distance' : result_distance, 'duration': result_time})
    except:
        print(loc)
```

```python
final_df_lis = pd.DataFrame(final)
lx_houses = lx_houses.join(final_df_lis[['distance', 'duration']])
```
