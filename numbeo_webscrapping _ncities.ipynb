```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import time
```

```python
cities_list=["Lisbon", 'Barcelona']
city= "Lisbon"
```

```python
##function to get data for certain city

def webscrab_items(cities):
    final = []
    time.sleep(4)
   
     ## requesting getting the information for the website. We can  define which list of cities we want
    for city in cities:
        url="https://www.numbeo.com/cost-of-living/in/" + city

        response =  requests.get(url)
        print(response.status_code)

        ##first look at the content
        soup = BeautifulSoup(response.content)
        
        ## find the table that is the parent tag of the tags I want 
        table_soup = soup.find("table", attrs={"class": "data_wide_table new_bar_table"})
        

        type(table_soup)

        ## LOOKING FOR ALL THE ITEMS FROM WHERE WE WANT THE PRICE

        item_catg = table_soup.find_all("tr")


        items_list = [item.text for item in item_catg]

        ## We need to treat the data. WE just want the info before the price
        param_priced=[]

        for i in items_list:
            try:
                #print(i)
                span = re.search(' [0-9]\S',i).span()[0]
                param_priced.append(i[:span]) 
                
            except:
                pass
        
        price_li = table_soup.find_all("span", attrs= {"class": "first_currency"})

        price_clean = [price.text.replace("\xa0", " ") for price in price_li]

        final.append({'param': param_priced, city : price_clean})

    return final
    



    
```

```python
data = webscrab_items(cities_list)
```

```python
bar = pd.DataFrame.from_dict(data[1])
```

```python
lis = pd.DataFrame.from_dict(data[0])
merged_df=pd.merge(bar,lis)
merged_df
```
